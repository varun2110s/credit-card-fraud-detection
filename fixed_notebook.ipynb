```python
fraud = df[df['class'] == 1]
normal = df[df['class'] == 0]
print(fraud.shape , normal.shape)
```

```python
plt.hist(fraud['amount'] , bins=50 , label='fraud')
plt.hist(normal['amount'] , bins=50 , label ='normal')
plt.legend()
plt.title("amount distribution")
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['Amount_scaled'] = scaler.fit_transform(df[['amount']])
df['Time_scaled'] = scaler.fit_transform(df[['Time']])
                
```

```python
print(df.head())
```

```python
df = df.drop(['amount' , 'Time'] , axis=1)
```

```python
print(df.columns.tolist()) # verifying the columns list 
```

```python
features = df.drop(['class'], axis = 1)
target = df['class']
```

```python
print(features.shape)
print(target.shape)
```

```python
# Splitting data into 80% train and 20% test
from sklearn.model_selection import train_test_split
features_train, features_test, target_train, target_test = train_test_split(features , target , test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
model =  LogisticRegression()
model.fit(features_train , target_train)
```

```python
predictions = model.predict(features_test)
```

```python
print(predictions[:10])
```

```python
from sklearn.metrics import classification_report , confusion_matrix 
print(classification_report( target_test, predictions))
```

```python
from imblearn.over_sampling import SMOTE
```

```python
smote = SMOTE(random_state=42)
features_train_smote , target_train_smote = smote.fit_resample(features_train, target_train)
```

```python
 print(target_train_smote.value_counts())
```

```python
model_smote = LogisticRegression()
```

```python
model_smote.fit(features_train_smote, target_train_smote)
```

```python
print(type(model_smote))
```

```python
predictions_smote = model_smote.predict(features_test)
```

```python
print(classification_report(target_test, predictions_smote))
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
model_rfc = RandomForestClassifier(n_jobs=-1, random_state=42)
model_rfc.fit(features_train_smote, target_train_smote)
```

```python
predictions_rfc = model_rfc.predict(features_test)
```

```python
print(classification_report(target_test , predictions_rfc))
```

```python
from xgboost import XGBClassifier
```

```python
model_xgb = XGBClassifier()
```

```python
model_xgb.fit(features_train_smote , target_train_smote)
```

```python
predictions_xgb = model_xgb.predict(features_test)
```

```python
print(classification_report(target_test , predictions_xgb))
```

```python
cm = confusion_matrix(target_test, predictions_rfc)
sns.heatmap(cm, annot=True, fmt='d', 
            xticklabels=['Normal','Fraud'], 
            yticklabels=['Normal','Fraud'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Random Forest')
plt.show()
```

```python
features_importance = pd.Series(model_rfc.feature_importances_,
                                index=features.columns)
features_importance.nlargest(10).plot(kind='barh')
plt.title('Top 10 Important Features')
plt.show()
```

```python
from sklearn.metrics import roc_auc_score, roc_curve

y_prob = model_rfc.predict_proba(features_test)[:,1]
print("AUC Score:", roc_auc_score(target_test, y_prob))

fpr, tpr, _ = roc_curve(target_test, y_prob)
plt.plot(fpr, tpr)
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Random Forest')
plt.show()
```
